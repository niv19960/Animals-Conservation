### Fill NaN values of Trend column

###### Because of the lack of knowledge about the "Trend" column, we chose to fill the NaN values with the neutral value - "Stable" (value = 1).
###### We don't want our variable to be depend on another variables, so we decided not to fill the missing values with the frequent value or according another parameters.

```python
animals_df.Trend.fillna(1, inplace=True)
```

```python
missing_status_df = animals_df[animals_df['Trend'].isna()]
missing_status_df
```

```python
animals_df.Trend.value_counts()
```

```python
mean_values = animals_df.groupby('Order')['Life Span (years)', 'Length (cm)', 'Weight (kg)',
                                         'Incubation Period (days)', 'Baby Carrying', 'Independent Age (days)',
                                         'Diet', 'Mating Behavior'].mean()
mean_values.head(10)
```

```python
animals_df.head(10)
```

### We filled missing continous variables data by mean

```python
for index, row in animals_df.iterrows():
    if math.isnan(row['Life Span (years)']):
        new_life_span = mean_values.loc[row['Order'], 'Life Span (years)']
        if new_life_span:
            animals_df.loc[index, 'Life Span (years)'] = new_life_span
```

```python
for index, row in animals_df.iterrows():
    
    if math.isnan(row['Life Span (years)']):
        new_life_span = mean_values.loc[row['Order'], 'Life Span (years)']
        if new_life_span:
            animals_df.loc[index, 'Life Span (years)'] = new_life_span
            
    if math.isnan(row['Length (cm)']):
        new_length = mean_values.loc[row['Order'], 'Length (cm)']
        if new_length:
            animals_df.loc[index, 'Length (cm)'] = new_length
            
    if math.isnan(row['Weight (kg)']):
        new_weight = mean_values.loc[row['Order'], 'Weight (kg)']
        if new_weight:
            animals_df.loc[index, 'Weight (kg)'] = new_weight
            
    if math.isnan(row['Incubation Period (days)']):
        new_incubation = mean_values.loc[row['Order'], 'Incubation Period (days)']
        if new_incubation:
            animals_df.loc[index, 'Incubation Period (days)'] = new_incubation
            
    if math.isnan(row['Baby Carrying']):
        new_baby = mean_values.loc[row['Order'], 'Baby Carrying']
        if new_baby:
            animals_df.loc[index, 'Baby Carrying'] = new_baby
            
    if math.isnan(row['Independent Age (days)']):
        new_age = mean_values.loc[row['Order'], 'Independent Age (days)']
        if new_age:
            animals_df.loc[index, 'Independent Age (days)'] = new_age
            
    if math.isnan(row['Diet']):
        new_diet = mean_values.loc[row['Order'], 'Diet']
        if new_diet:
            animals_df.loc[index, 'Diet'] = np.rint(new_diet)
            
    if math.isnan(row['Mating Behavior']):
        new_mate = mean_values.loc[row['Order'], 'Mating Behavior']
        if new_mate:
            animals_df.loc[index, 'Mating Behavior'] = np.rint(new_mate)

# Life Span (years), Length (cm), Weight (kg), Incubation Period (days),
# Baby Carrying, Independent Age (days), Diet, Mating Behavior
# mean_values, mode_values
```

```python
animals_df.info()
```

```python
values = {'Life Span (years)':animals_df['Life Span (years)'].mean(),
         'Length (cm)':animals_df['Length (cm)'].mean(),
         'Mating Behavior':animals_df['Mating Behavior'].mode()[0],
         'Incubation Period (days)': animals_df['Incubation Period (days)'].mean(),
         'Independent Age (days)': animals_df['Independent Age (days)'].mean()}
animals_df.fillna(value=values, inplace=True)
```

```python
animals_df.info()
```

### Convert object variables into Categorical variables

```python
animals_df['Trend'] = animals_df['Trend'].astype('category').cat.codes
animals_df['Diet'] = animals_df['Diet'].astype('category').cat.codes
animals_df['Mating Behavior'] = animals_df['Mating Behavior'].astype('category').cat.codes

animals_df['Class'] = animals_df['Class'].astype('category').cat.codes
animals_df['Order'] = animals_df['Order'].astype('category').cat.codes
```

```python
animals_df.to_csv("ANIMALS DATA - final.csv", index=False)
```
